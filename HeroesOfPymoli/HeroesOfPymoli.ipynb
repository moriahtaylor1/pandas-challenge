### Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
df2 = df.drop_duplicates(subset="username")
#count of male players
count_male = len(df2[df2["gender"]=="Male"])
#count of female players
count_female = len(df2[df2["gender"]=="Female"])
#count of other/non-disclosed
count_other = len(df2) - count_male - count_female
#perc of male players
##calculate percentage
perc_male = count_male / len(df2)
##format as string
formatted_perc_male = perc_format(perc_male)
#perc of female players
##calculate percentage
perc_female = count_female / len(df2)
##format as string
formatted_perc_female = perc_format(perc_female)
#perc of other/non-disclosed
##calculate percentage
perc_other = count_other / len(df2)
##format as string
formatted_perc_other = perc_format(perc_other)
#print using tabulate function
table3 = [["Male", count_male, formatted_perc_male],
         ["Female", count_female, formatted_perc_female],
         ["Other/Non-Disclosed", count_other, formatted_perc_other]]
print(tabulate(table3, headers = ["Gender", "Count", "Percentage"]))
```


### Purchasing Analysis by Gender

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#create dataframes for each gender identification
male_df = df[df["gender"]=="Male"]
female_df = df[df["gender"]=="Female"]
others_df = df[-df["gender"].isin(["Male", "Female"])]



#calculations for males
male_purchase_count = len(male_df)
male_avg_purchase = male_df["price"].mean()
male_total = male_df["price"].sum()
male_avg_per_person = male_total / count_male

#calculations for females
female_purchase_count = len(female_df)
female_avg_purchase = female_df["price"].mean()
female_total = female_df["price"].sum()
female_avg_per_person = female_total / count_female

#calculations for others
others_purchase_count = len(others_df)
others_avg_purchase = others_df["price"].mean()
others_total = others_df["price"].sum()
others_avg_per_person = others_total / count_other

table4 = [["Gender", "", "", ""],
         ["Female", female_purchase_count, money_format(female_avg_purchase), money_format(female_total), money_format(female_avg_per_person)],
         ["Male", male_purchase_count, money_format(male_avg_purchase), money_format(male_total), money_format(male_avg_per_person)],
         ["Other / Non-Disclosed", others_purchase_count, money_format(others_avg_purchase), money_format(others_total), money_format(others_avg_per_person)]]
print(tabulate(table4, headers=["", "Purchase Count", "Avg Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]))
```
